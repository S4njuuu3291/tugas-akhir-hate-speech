```python
import numpy as np

X = [1,3,5]
Y = [2,3,4]

X_avg = np.average(X)
Y_avg = np.average(Y)

X_diff = X - X_avg
Y_diff = Y - Y_avg

atas = X_diff @ Y_diff
atas
```

```python
kiri = 0
kanan = 0
for i,j in zip(X_diff,Y_diff):
    kiri+= i**2
    kanan+= j**2

bawah = np.sqrt(kiri*kanan)

korelasi = atas/bawah
korelasi
```

```python
A = [1,0,1,0,1]
B = [1,1,0,0,1]

data_set = [A,B]

avgs = [np.average(data) for data in data_set]

diffs = [data - avg for avg,data in zip(avgs,data_set)]

atas = []
for i,diff in enumerate(diffs):
    if i < len(diffs)-1:
        atas.append(diffs[i]@diffs[i+1])

sum_kuadrat_a = np.sum(diffs[0]**2)
sum_kuadrat_b = np.sum(diffs[1]**2)

bawah = np.sqrt(sum_kuadrat_a*sum_kuadrat_b)

atas/bawah
```
