```python
import numpy as np
TFIDFkalimat1 = np.array([0.25, 0.25, 0.17, 0.25, 0.36])

bobot = np.array([1.0, 1.5, 0.5, 1.0, 1.2])

hasilLogreg = [TFIDFkalimat1[i] * bobot[i] for i in range(len(TFIDFkalimat1))]

sum(hasilLogreg)

def sigmoid(x): return 1/(1+np.exp(-x))

print("Hasil LogReg sebelum sigmoid:", sum(hasilLogreg))
print("Hasil LogReg setelah sigmoid:", sigmoid(sum(hasilLogreg)))

# fungsinya untuk memetakan hasil logits ke rentang 0-1, yang bisa diinterpretasikan sebagai probabilitas.
```
