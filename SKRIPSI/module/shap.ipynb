```python
import numpy as np

X = [[20,2],
     [25,5],
     [30,3],
     [35,8],
     [40,6]]

y = [30,45,50,60,65]

X = np.array(X)
X = np.c_[np.ones(X.shape[0]), X]
y = np.array(y)
print(X.shape,y.shape)

X.T @ y

print("Rule:",np.linalg.inv((X.T @ X) )@ (X.T @ y))
```

```python
import numpy as np

# Matriks Query (Q) - Ukuran [3, 4]
Q = np.array([
    [2, 0, 1, 3],
    [0, 4, 1, 1],
    [3, 1, 0, 2]
], dtype=np.float32)

# Matriks Key (K) - Ukuran [3, 4]
K = np.array([
    [1, 3, 0, 2],
    [2, 0, 4, 1],
    [0, 1, 1, 3]
], dtype=np.float32)

# Matriks Value (V) - Ukuran [3, 4]
V = np.array([
    [3, 1, 4, 2],
    [1, 5, 0, 3],
    [2, 2, 6, 1]
], dtype=np.float32)

# --- VERIFIKASI DIMENSI ---
print("Dimensi Q:", Q.shape)  # Output: (3, 4)
print("Dimensi K:", K.shape)  # Output: (3, 4)
print("Dimensi V:", V.shape)  # Output: (3, 4)

Q,K,V
```

```python
def softmax(arr):
    hasil = []
    for x in arr:
        hasil.append(np.exp(x)/np.sum(np.exp(arr)))

    return hasil

arr = [4,1,-2]
softmax(arr)
```

```python
Q = np.array(Q)
K = np.array(K)
V = np.array(V)

softmax((Q @ K.T)/np.sqrt(4)) @ V
```
