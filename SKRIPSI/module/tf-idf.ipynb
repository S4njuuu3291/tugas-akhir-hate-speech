```python
import numpy as np

kalimat1 = "saya benci kamu" 
kalimat2 = "saya suka kamu"
kalimat3 = "dia benci kamu"
kalimat4 = "dia suka kamu"

kataUnik = ["saya","benci", "suka", "kamu", "dia"]

tf1 = [1,1,0,1,0] # frekuensi kata dalam kalimat 1
tf2 = [1,0,1,1,0] # frekuensi kata dalam kalimat 2
tf3 = [0,1,0,1,1] # frekuensi kata dalam kalimat 3
tf4 = [0,0,1,1,1] # frekuensi kata dalam kalimat 4

sayaIdf = np.log(2/2)
benciIdf = np.log(2/1)
sukaIdf = np.log(2/1)
kamuIdf = np.log(2/2)
diaIdf = np.log(2/2)

tfKalimat1 = [1*sayaIdf, 1*benciIdf, 0*sukaIdf, 1*kamuIdf, 0*diaIdf]
tfKalimat2 = [1*sayaIdf, 0*benciIdf, 1*sukaIdf, 1*kamuIdf, 0*diaIdf]
tfKalimat3 = [0*sayaIdf, 1*benciIdf, 0*sukaIdf, 1*kamuIdf, 1*diaIdf]
tfKalimat4 = [0*sayaIdf, 0*benciIdf, 1*sukaIdf, 1*kamuIdf, 1*diaIdf]

print("TF-IDF Kalimat 1:", tfKalimat1)
print("TF-IDF Kalimat 2:", tfKalimat2)
print("TF-IDF Kalimat 3:", tfKalimat3)
print("TF-IDF Kalimat 4:", tfKalimat4)
```

```python
kalimat1TfIdf = [tf1[i] * [sayaIdf, benciIdf, sukaIdf, kamuIdf, diaIdf][i] for i in range(len(kataUnik))]
kalimat2TfIdf = [tf2[i] * [sayaIdf, benciIdf, sukaIdf, kamuIdf, diaIdf][i] for i in range(len(kataUnik))]
kalimat3TfIdf = [tf3[i] * [sayaIdf, benciIdf, sukaIdf, kamuIdf, diaIdf][i] for i in range(len(kataUnik))]
kalimat4TfIdf = [tf4[i] * [sayaIdf, benciIdf, sukaIdf, kamuIdf, diaIdf][i] for i in range(len(kataUnik))]

print("TF-IDF Kalimat 1:", kalimat1TfIdf)
print("TF-IDF Kalimat 2:", kalimat2TfIdf)
print("TF-IDF Kalimat 3:", kalimat3TfIdf)
print("TF-IDF Kalimat 4:", kalimat4TfIdf)
```
