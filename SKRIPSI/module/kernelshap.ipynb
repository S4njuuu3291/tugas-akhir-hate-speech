```python
import matplotlib.pyplot as plt

def factorial(x):
    value = 1
    for i in range(1,x+1):
        value = value*i
    return value

x = 5
print(factorial(x))

def combination(x,y):
    return (factorial(x)/(factorial(y)*factorial(x-y)))

print(combination(3,1))

def kernel_weight(M,k):
    if k == 0 or k == M:
        return 1e6
    return (M-1)/(combination(M,k)*k*(M-k))

data = []
M = 5
for i in range(M,0,-1):
    data.append(kernel_weight(M,i))
    print(i)

print(data)
plt.plot(data)
plt.show()
```
